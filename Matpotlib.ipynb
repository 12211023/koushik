```python
# Importing matplotlib library
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Simple Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title('Simple Line Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Horizontal Bar Plot
plt.barh(categories, values)
plt.title('Horizontal Bar Plot')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

```python
# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=30)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pie Chart
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# Box Plot
data = [np.random.rand(50) for _ in range(4)]
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Violin Plot
data = [np.random.rand(50) for _ in range(4)]
plt.violinplot(data)
plt.title('Violin Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Error Bar Plot
x = np.linspace(0, 10, 10)
y = np.sin(x)
yerr = 0.1
plt.errorbar(x, y, yerr=yerr, fmt='o')
plt.title('Error Bar Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# Multiple Lines in One Plot
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.title('Multiple Lines in One Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Subplots
x = np.linspace(0, 10, 100)
fig, axs = plt.subplots(2)
axs[0].plot(x, np.sin(x))
axs[0].set_title('sin(x)')
axs[1].plot(x, np.cos(x))
axs[1].set_title('cos(x)')
plt.show()
```

```python
# Twin Axes Plot
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
ax1.plot(x, y1, 'g-')
ax2.plot(x, y2, 'b-')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)', color='g')
ax2.set_ylabel('cos(x)', color='b')
plt.title('Twin Axes Plot')
plt.show()
```

```python
# Logarithmic Scale Plot
x = np.logspace(0.1, 2, 100)
y = np.exp(x)
plt.plot(x, y)
plt.xscale('log')
plt.yscale('log')
plt.title('Logarithmic Scale Plot')
plt.xlabel('log(x)')
plt.ylabel('log(y)')
plt.show()
```

```python
# Polar Plot
theta = np.linspace(0, 2 * np.pi, 100)
r = np.abs(np.sin(theta))
plt.polar(theta, r)
plt.title('Polar Plot')
plt.show()
```
