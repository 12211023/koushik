```python
# Importing matplotlib library
import matplotlib.pyplot as plt
import numpy as np

```

```python
# Simple Line Plot with Customizations
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color='red', linestyle='--', linewidth=2)
plt.title('Customized Line Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# Grouped Bar Plot
categories = ['A', 'B', 'C', 'D']
values1 = [10, 20, 15, 25]
values2 = [5, 15, 10, 20]
bar_width = 0.35
index = np.arange(len(categories))
plt.bar(index, values1, bar_width, label='Group 1')
plt.bar(index + bar_width, values2, bar_width, label='Group 2')
plt.title('Grouped Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.xticks(index + bar_width / 2, categories)
plt.legend()
plt.show()
```

```python
# Scatter Plot with Color and Size
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.title('Scatter Plot with Color and Size')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.show()
```

```python
# Horizontal Bar Plot with Colors
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
colors = ['red', 'blue', 'green', 'purple']
plt.barh(categories, values, color=colors)
plt.title('Horizontal Bar Plot with Colors')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
```

```python
# Stacked Bar Plot
categories = ['A', 'B', 'C', 'D']
values1 = [10, 20, 15, 25]
values2 = [5, 10, 5, 10]
plt.bar(categories, values1, label='Group 1')
plt.bar(categories, values2, bottom=values1, label='Group 2')
plt.title('Stacked Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
# Multiple Histograms
data1 = np.random.randn(1000)
data2 = np.random.randn(1000)
plt.hist(data1, bins=30, alpha=0.5, label='Data 1')
plt.hist(data2, bins=30, alpha=0.5, label='Data 2')
plt.title('Multiple Histograms')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Donut Chart
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', wedgeprops=dict(width=0.3))
plt.title('Donut Chart')
plt.show()
```

```python
# Box Plot with Notches
data = [np.random.rand(50) for _ in range(4)]
plt.boxplot(data, notch=True)
plt.title('Box Plot with Notches')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Horizontal Error Bar Plot
x = np.linspace(0, 10, 10)
y = np.sin(x)
xerr = 0.1
plt.errorbar(x, y, xerr=xerr, fmt='o')
plt.title('Horizontal Error Bar Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# Filled Line Plot
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.fill_between(x, y1, y2, color='lightblue', alpha=0.5)
plt.title('Filled Line Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Subplots with Shared Axes
x = np.linspace(0, 10, 100)
fig, axs = plt.subplots(2, sharex=True)
axs[0].plot(x, np.sin(x))
axs[0].set_title('sin(x)')
axs[1].plot(x, np.cos(x))
axs[1].set_title('cos(x)')
plt.show()
```

```python
# Annotating Multiple Points
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.annotate('max', xy=(1.5, 1), xytext=(2, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('min', xy=(4.5, -1), xytext=(5, -1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.title('Annotating Multiple Points')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# Density Plot
data = np.random.randn(1000)
density, bins, _ = plt.hist(data, bins=30, density=True, alpha=0.5)
count, division = np.histogram(data, bins=bins)
plt.plot(0.5 * (bins[1:] + bins[:-1]), density, '-')
plt.title('Density Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
# Customizing Ticks
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.xticks([0, 2, 4, 6, 8, 10])
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.title('Customizing Ticks')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# Customized Legends
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), label='sin(x)', linestyle='--')
plt.plot(x, np.cos(x), label='cos(x)', linestyle=':')
plt.title('Customized Legends')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right', fontsize='large', frameon=True, shadow=True)
plt.show()
```

```python
# Changing Spines
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_position(('outward', 10))
ax.spines['bottom'].set_position(('outward', 10))
plt.title('Changing Spines')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# Shaded Error Bar Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
yerr = 0.1 + 0.1 * np.sqrt(x)
plt.plot(x, y)
plt.fill_between(x, y - yerr, y + yerr, alpha=0.3)
plt.title('Shaded Error Bar Plot')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# Line Plot with Markers
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, marker='o')
plt.title('Line Plot with Markers')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
# Bubble Plot
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
plt.scatter(x, y, s=sizes, alpha=0.5)
plt.title('Bubble Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
